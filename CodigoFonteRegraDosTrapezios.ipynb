# Regra dos Trapézios

```python
!pip install -q mplcyberpunk
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import mplcyberpunk
plt.style.use("cyberpunk")
```

```python
def RegraDosTrapezios(vetor, numParticoes):
    h = (vetor[-1][0] - vetor[0][0])/numParticoes
    h = round(h,10)
    soma = 0

    for i in range(1,len(vetor)):
        soma += (h*(vetor[i][1] + vetor[i-1][1]))/2

    return soma
```
